# **OLIST ECOMMERCE ANALYSIS**

## **Objective**

This project aims to build predictive machine learning models using e-commerce data to analyze and forecast product pricing and delivery behavior. By applying regression and classification algorithms, the goal is to:

Predict product price based on key features like weight, dimensions, installment methods, and freight value

Analyze customer and payment patterns

Classify order outcomes (e.g., delivered, canceled) using order and customer attributes

Identify trends in product categories, states, and seller behavior


This allows businesses to make data-driven decisions for pricing strategy, delivery performance, and customer engagement

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import optuna
```

```python
# LinearRegression using GridSearchCV

lr = LinearRegression()

# Hyperparameter grid (LinearRegression has limited tunable params)
param_grid = {"fit_intercept": [True, False]}

# GridSearchCV
grid_search = GridSearchCV(lr, param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train_numeric, y_train)

# Best model
best_lr = grid_search.best_estimator_
print("Best Params:", grid_search.best_params_)

# Predictions
y_pred = best_lr.predict(X_test_numeric)

# Metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\n Evaluation Metrics (Best Linear Regression):")
print(f"   MSE : {mse:.2f}")
print(f"   RMSE: {rmse:.2f}")
print(f"   R²  : {r2:.4f}")
```

```python
# KNeighborsRegressor with GridSerachCV

# Define KNN model
knn = KNeighborsRegressor()

# Step 2: Create a smaller param grid for faster tuning
param_grid = {'n_neighbors': [3],'p': [1, 2]}      # 1: Manhattan, 2: Euclidean


# Grid Search with 3-fold CV
grid_knn = GridSearchCV(knn, param_grid, cv=3, scoring="neg_mean_squared_error", n_jobs=1)
grid_knn.fit(X_train_numeric, y_train)

#  Predict on test set
y_pred = grid_knn.predict(X_test_numeric)

#  Metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)


print("Best Parameters:", grid_knn.best_params_)
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
# Define Parameter Ridge

ridge_model=Ridge()
param_grid = {"alpha": [0.001, 0.01, 0.1, 1, 10, 100]}
grid_search = GridSearchCV(ridge_model, param_grid, cv=5, scoring="neg_mean_squared_error", n_jobs=-1)
grid_search.fit(X_train_numeric, y_train)

# Best model
best_ridge = grid_search.best_estimator_
print("Best Params:", grid_search.best_params_)


y_pred = best_ridge.predict(X_test_numeric)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\n Evaluation Metrics (Best Ridge Regression):")
print(f"   MSE : {mse:.2f}")
print(f"   RMSE: {rmse:.2f}")
print(f"   R²  : {r2:.4f}")
```

```python
lasso_model=Lasso()
param_grid={"alpha":[0.001,0.01,0.1,1,10,100]}
grid_search= GridSearchCV(lasso_model,param_grid,cv=5,scoring="neg_mean_squared_error")
grid_search.fit(X_train_numeric,y_train)
best_lasso=grid_search.best_estimator_
print("best_params:",grid_search.best_params_)
y_pred=best_lasso.predict(X_test_numeric)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)

print("\n Evaluation Metrices (Best Lasso Regression)")
print(f" MSE:{mse:.2f}")
print(f" RMSE:{rmse:.2f}")
print(f" R2:{r2:.4f}")
```

```python
# Define Random Forest

rf_model=RandomForestRegressor()
param_rf={"n_estimators":[50,100],"max_depth":[None,10,20],"min_samples_split":[2,5]}
random_search=RandomizedSearchCV(rf_model,param_distributions=param_rf,n_iter=5,cv=3,scoring="neg_mean_squared_error")
random_search.fit(X_train_numeric,y_train)
best_rf=random_search.best_estimator_
print("best_params:",random_search.best_params_)
y_pred=best_rf.predict(X_test_numeric)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)

print("\n Evaluation Metices:(Best Random Forest Regression)")
print(f" MSE:{mse:.2f}")
print(f" RMSE:{rmse:.2f}")
print(f" R2:{r2:.4f}")
```

```python
# Define GradientBoosting

gb_model=GradientBoostingRegressor()
param_gb={"n_estimators":[50,70],"learning_rate":[0.01,0.1,1],"max_depth":[3,4]}
random_search=RandomizedSearchCV(gb_model,param_distributions=param_gb,n_iter=5,cv=3,scoring="neg_mean_squared_error")
random_search.fit(X_train_numeric,y_train)
best_gb=random_search.best_estimator_
print("best_params:",random_search.best_params_)
y_pred=best_gb.predict(X_test_numeric)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)

print("\n Evaluation Metrics:(Best Gradient Boosting Regression)")
print(f" MSE:{mse:.2f}")
print(f" RMSE:{rmse:.2f}")
print(f" R2:{r2:.4f}")
```

```python
# Define XGBoost

xgboost_model=XGBRegressor()
param_xgb={"n_estimators":[50,100],"learning_rate":[0.01,0.1,1],"max_depth":[3,4]}
random_search=RandomizedSearchCV(xgboost_model,param_distributions=param_xgb,n_iter=5,cv=4,scoring="neg_mean_squared_error")
random_search.fit(X_train_numeric,y_train)
best_xgb=random_search.best_estimator_
print("best_params:",random_search.best_params_)
y_pred=best_xgb.predict(X_test_numeric)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)

print("\n Evaluation Metrics:(Best XGBoost Regression)")
print(f" MSE:{mse:.2f}")
print(f" RMSE:{rmse:.2f}")
print(f" R2:{r2:.4f}")
```

```python
from sklearn.linear_model import LogisticRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import optuna
```

```python
df['status_binary'] = df['order_status'].apply(lambda x: 1 if x == 'delivered' else 0)
```

```python
x=df.select_dtypes(include=['int64', 'float64']).drop(columns=["review_score", "payment_sequential", "payment_installments", "payment_value", "seller_zip_code_prefix", "geolocation_zip_code_prefix", "geolocation_lat", "geolocation_lng", "customer_zip_code_prefix"])
y=df["status_binary"]
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 1. Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# 2. Initialize LabelEncoder
le = LabelEncoder()

# 3. Apply LabelEncoder to each categorical column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
from sklearn.model_selection import train_test_split

x=df.drop(columns=["review_score", "payment_sequential", "payment_installments", "payment_value", "seller_zip_code_prefix", "geolocation_zip_code_prefix", "geolocation_lat", "geolocation_lng", "customer_zip_code_prefix", "status_binary"])
y=df["status_binary"]

X_train, X_test_cate, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler=StandardScaler()
X_train_scaler=scaler.fit_transform(X_train)
X_test_scaler=scaler.transform(X_test_cate)
```

```python
# Logistic Regression

# Define param grid
param_random = {
    "C": [0.01, 0.1, 1],
    "penalty": ["l1"],
    "solver": ["liblinear"]
}

# Define model
lr = LogisticRegression(max_iter=300, random_state=42)

# RandomSearchCV
from sklearn.model_selection import RandomizedSearchCV
logreg_random = RandomizedSearchCV(lr, param_random, cv=3, scoring="accuracy", n_jobs=-1)
logreg_random.fit(X_train, y_train)

# Predict
logreg_predict = logreg_random.predict(X_test_cate)

# Evaluate
print("Logistic Regression Results")
print("Best Parameters:", logreg_random.best_params_)
print("Accuracy:", round(accuracy_score(y_test, logreg_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, logreg_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, logreg_predict))
print("Classification Report:\n", classification_report(y_test, logreg_predict))
```

```python
# Define parameter grid
knn_params = {'n_neighbors': [3, 5, 7]}

# Create GridSearchCV
knn_grid = GridSearchCV(KNeighborsClassifier(), knn_params, cv=3, scoring='accuracy')
knn_grid.fit(X_train, y_train)

# Predict
knn_predict = knn_grid.predict(X_test_cate)

# Evaluate
print("KNN")
print("Best Params:", knn_grid.best_params_)
print("Accuracy:", round(accuracy_score(y_test, knn_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, knn_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, knn_predict))
print("Classification Report:\n", classification_report(y_test, knn_predict))
```

```python
rfc = RandomForestClassifier()

# Define hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Grid Search with 3-fold CV
grid_rfc = GridSearchCV(rfc, param_grid, cv=3, scoring="accuracy", n_jobs=-1)
grid_rfc.fit(X_train, y_train)

# Predict
rfc_predict = grid_rfc.predict(X_test_cate)

# Evaluation
print("Random Forest Classifier")
print("Best Parameters:", grid_rfc.best_params_)
print("Accuracy:", round(accuracy_score(y_test, rfc_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, rfc_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, rfc_predict))
print("Classification Report:\n", classification_report(y_test, rfc_predict))
```

```python
xgb = XGBClassifier()

# Hyperparameter grid
param_grid = {'n_estimators': [100, 200], 'max_depth': [3, 5],'learning_rate': [0.05, 0.1],'subsample': [0.8, 1]}

# GridSearchCV
grid_xgb = GridSearchCV(estimator=xgb, param_grid=param_grid,
                        cv=3, scoring='accuracy', n_jobs=-1)

# Fit the model
grid_xgb.fit(X_train, y_train)

# Predict
xgb_predict = grid_xgb.predict(X_test_cate)

# Evaluation
print("XGBoost Classifier")
print("Best Parameters:", grid_xgb.best_params_)
print("Accuracy:", round(accuracy_score(y_test, xgb_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, xgb_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, xgb_predict))
print("Classification Report:\n", classification_report(y_test, xgb_predict))
```

```python
# GradientBoostingClassifier using GridSearchCV

# Define model
gb = GradientBoostingClassifier()

# Hyperparameter grid
param_grid = {'n_estimators': [100, 150],'learning_rate': [0.05, 0.1],'max_depth': [3, 5]}

# Grid Search
grid_gb = GridSearchCV(estimator=gb, param_grid=param_grid,
                        cv=3, scoring='accuracy', n_jobs=-1)

# Fit the model
grid_gb.fit(X_train, y_train)

# Predict
gb_predict = grid_gb.predict(X_test_cate)

# Evaluation
print("Gradient Boosting Classifier")
print("Best Parameters:", grid_gb.best_params_)
print("Accuracy:", round(accuracy_score(y_test, gb_predict) * 100, 2), "%")
print("F1 Score:", round(f1_score(y_test, gb_predict, average='macro'), 3))
print("Confusion Matrix:\n", confusion_matrix(y_test, gb_predict))
print("Classification Report:\n", classification_report(y_test, gb_predict))
```
